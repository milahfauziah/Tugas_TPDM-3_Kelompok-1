```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#panggil dataset
data = pd.read_csv('instagram.csv')
data.head()
```

```python
import re
#buat fungsi case folding 
def casefolding(content):
    content = content.lower()            #ubah kalimat jadi huruf kecil
    content = re.sub(r'https?://\S+|www\.\S+','', content) # hapus url
    content = re.sub(r'[-+]?[0-9]+', '', content)  # hapus angka
    content = re.sub(r'[^\w\s]', '', content)  # hapus tanda baca
    content = content.strip()
    return content
```

```python
raw_sample = data['content'].iloc[2]
case_folding = casefolding(raw_sample)

print('Raw data\t :', raw_sample)
print('Case Folding\t:', case_folding)
```

```python
#Tahap normalisasi kata
key_norm = pd.read_csv('key_norm.csv')

def content_normalize(content):
    content = ' '.join([key_norm[key_norm['singkat'] == word]['hasil'].values[0]
                    if (key_norm['singkat'] == word).any()
                     else word for word in content.split()
                    ])
    
    content = str.lower(content)
    return content
```

```python
#membandingkan data
raw_data = data['content'].iloc[2]
word_normal = content_normalize(case_folding)

print('Raw data\t : ', raw_data)
print('Word Normalize\t : ', word_normal)
```

```python
#filtering (stopword removal)
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

stopwords_ind = stopwords.words('indonesian')
```

```python
len(stopwords_ind)
```

```python
#daftar stopword dari nltk
stopwords_ind
```

```python
#fungsi stopword removal
more_stopword = ['dech']
stopwords_ind = stopwords_ind + more_stopword

def remove_stop_word(content):
    clean_words = []
    content = content.split()
    for word in content:
        if word not in stopwords_ind:
            clean_words.append(word)
    return " ".join(clean_words)
```

```python
raw_sample = data['content'].iloc[14]
case_folding = casefolding(raw_sample)
stopword_removal = remove_stop_word(case_folding)

print('Raw data\t : ', raw_data)
print('Case Folding\t : ', case_folding)
print('Stopword Removal\t : ', stopword_removal)
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemming(content):
    content = stemmer.stem(content)
    return content
```

```python
raw_sample = data['content'].iloc[14]
case_folding = casefolding(raw_sample)
stopword_removal = remove_stop_word(case_folding)
content_stemming = stemming(stopword_removal)

print('Raw data\t : ', raw_data)
print('Case Folding\t : ', case_folding)
print('Stopword Removal\t : ', stopword_removal)
print('Stemming\t : ', content_stemming)
```

```python
#text preprocessing pipeline
#gabung seluruh langkah text preprocessing

def text_preprocessing_process(content):
    content = casefolding(content)
    content = content_normalize(content)
    content = remove_stop_word(content)
    content = stemming(content)
    return content
```

```python
%%time
data['clean_teks']=data['content'].apply(text_preprocessing_process)
```

```python
##feature extraction(TF-IDF & N-gram)
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer

vec_TF_IDF = TfidfVectorizer(ngram_range=(1,1))
vec_TF_IDF.fit(x)

x_tf_idf = vec_TF_IDF.transform(x)
pickle.dump(vec_TF_IDF.vocabulary_,open("feature_tf-idf.sav", "wb"))
```

```python
#feature selection
x_train = np.array(data_tabular_tf_idf)
y_train = np.array(y)
```
